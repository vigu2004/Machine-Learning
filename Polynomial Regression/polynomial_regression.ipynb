# Polynomial Regression


```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
```

```python
dataset = pd.read_csv("Position_Salaries.csv")
X = dataset.iloc[:,1:-1]
y = dataset.iloc[:,-1]
```

```python
print(X)
print(y)
```

# Training Linear Regression and then Polynomial Regression 

```python
from sklearn.linear_model import LinearRegression

linear_regressor = LinearRegression()
linear_regressor.fit(X,y)
```

# Create new matrix of features with x , xsquare , xcube then use linear regression on that new matrix

```python
from sklearn.preprocessing import PolynomialFeatures

pf = PolynomialFeatures(degree=4)

X_poly = pf.fit_transform(X)
```

```python
print(X_poly)
```

```python
polynomial_regressor = LinearRegression()
polynomial_regressor.fit(X_poly,y)
```

```python
plt.scatter(X,y,color= "red")
plt.plot(X,linear_regressor.predict(X),color="blue")
plt.title("Linear Regression Visualization")
plt.xlabel("Independent Variable")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(X,y,color= "red")
plt.plot(X,polynomial_regressor.predict(X_poly),color="blue")
plt.title("Polynomial Regression Visualization")
plt.xlabel("Independent Variable")
plt.ylabel("Salary")
plt.show()
```

```python
linear_regressor.predict([[6.5]])
```

```python
polynomial_regressor.predict(pf.fit_transform([[6.5]]))
```
