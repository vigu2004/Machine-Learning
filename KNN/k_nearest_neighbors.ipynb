# K-Nearest Neighbors (K-NN)

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a new directed graph with increased spacing for the center node
G_spaced = nx.DiGraph()

# Define nodes and edges for the simpler mind map with increased spacing
nodes_spaced = {
    "Navigation System": ["Ultrasound", "Infrared", "GPS"],
    "Ultrasound": ["Advantages (US)", "Disadvantages (US)"],
    "Infrared": ["Advantages (IR)", "Disadvantages (IR)"],
    "GPS": ["Advantages (GPS)", "Disadvantages (GPS)"],
    "Advantages (US)": ["Accurate distance measurement", "Effective in detecting obstacles"],
    "Disadvantages (US)": ["Limited range", "Potential interference"],
    "Advantages (IR)": ["Good for detecting obstacles", "Can work in low light"],
    "Disadvantages (IR)": ["Limited range", "Susceptible to interference"],
    "Advantages (GPS)": ["Accurate location information", "Useful for outdoor navigation"],
    "Disadvantages (GPS)": ["Limited to outdoor use", "High battery consumption"]
}

# Add nodes and edges to the spaced graph
for parent, children in nodes_spaced.items():
    for child in children:
        G_spaced.add_edge(parent, child)

# Define a layout for the spaced graph with increased spacing for the center node
pos_spaced = nx.spring_layout(G_spaced, seed=42, k=0.2)

# Draw the spaced graph
plt.figure(figsize=(14, 10))
nx.draw(G_spaced, pos_spaced, with_labels=True, node_size=3500, node_color="lightblue", font_size=10, font_weight="bold", font_color="black", edge_color="gray", width=1.5, arrows=True, arrowstyle="-|>", arrowsize=15)
plt.title("Navigation System")
plt.show()
```

## Importing the dataset

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

## Training the K-NN model on the Training set

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

## Predicting a new result

```python
print(classifier.predict(sc.transform([[30,87000]])))
```

## Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

## Visualising the Training set results

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 1),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 1))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('K-NN (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

## Visualising the Test set results

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_test), y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 1),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 1))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('K-NN (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
